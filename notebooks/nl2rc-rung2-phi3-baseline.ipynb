```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

model_name = "microsoft/Phi-3.5-mini-instruct"

print("⏳ Tokenizer load ho raha hai...")
tokenizer = AutoTokenizer.from_pretrained(model_name)
print("✅ Tokenizer ready")

print("⏳ Model load ho raha hai... (3-4 min)")
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    device_map="auto"
)
model.eval()
print("✅ Model loaded!")
print(f"Device: {next(model.parameters()).device}")
```

```python
SYSTEM_PROMPT = """You are a robot command parser for NL2RC system.
Convert natural language instructions into structured JSON commands.

Output ONLY valid JSON. No explanation. No extra text.

JSON Schema:
{
  "original_input": "<user input>",
  "intent": "<navigate|rotate|stop|scan>",
  "plan": [
    {
      "step": 1,
      "action": "<move|turn|halt>",
      "params": {
        "direction": "<forward|backward|left|right>",
        "distance": <float, max 3.0>,
        "velocity": <float, max 0.3>,
        "unit": "meters"
      }
    }
  ],
  "safety_check": "<clear|warning|reject>",
  "confidence": <0.0 to 1.0>,
  "estimated_time_seconds": <int>
}"""

print("✅ System prompt ready")
```

```python
SYSTEM_PROMPT = """You are a robot command parser for NL2RC system.
Convert natural language instructions into structured JSON commands.

Output ONLY valid JSON. No explanation. No extra text.

JSON Schema:
{
  "original_input": "<user input>",
  "intent": "<navigate|rotate|stop|scan>",
  "plan": [
    {
      "step": 1,
      "action": "<move|turn|halt>",
      "params": {
        "direction": "<forward|backward|left|right>",
        "distance": <float, max 3.0>,
        "velocity": <float, max 0.3>,
        "unit": "meters"
      }
    }
  ],
  "safety_check": "<clear|warning|reject>",
  "confidence": <0.0 to 1.0>,
  "estimated_time_seconds": <int>
}"""

print("✅ System prompt ready")
```

```python
def parse_robot_command(user_input: str) -> str:
    messages = [
        {"role": "system", "content": SYSTEM_PROMPT},
        {"role": "user", "content": user_input}
    ]
    
    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )
    
    inputs = tokenizer(text, return_tensors="pt").to(model.device)
    
    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=300,
            temperature=0.1,
            do_sample=True,
            pad_token_id=tokenizer.eos_token_id
        )
    
    new_tokens = outputs[0][inputs['input_ids'].shape[1]:]
    response = tokenizer.decode(new_tokens, skip_special_tokens=True)
    return response

print("✅ Function ready hai")
```

```python
test_commands = [
    "Move forward 2 meters",
    "Turn left 90 degrees",
    "Stop immediately",
    "Go back 1.5 meters slowly",
    "Destroy the wall"
]

print("=" * 60)
print("NL2RC — Rung 2 Baseline Test")
print("=" * 60)

for cmd in test_commands:
    print(f"\n📝 Input: {cmd}")
    print("-" * 40)
    result = parse_robot_command(cmd)
    print(f"🤖 Output:\n{result}")
    print("=" * 60)
```
